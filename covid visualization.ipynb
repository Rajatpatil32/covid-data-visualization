```python
import covid
import matplotlib.pyplot as plt
```

```python
cov = covid.Covid()
name = input("enter country name that u want to see covid graph = ")
virusdata = cov.get_status_by_country_name(name)
active = virusdata['active']
recover = virusdata['recovered']
deaths = virusdata['deaths']
plt.pie([active,recover,deaths],
        labels = ['active','recovered','deaths'],
        colors = ['g','b','y'],
        explode = (0.5,0.3,0.2),
        startangle = 180,autopct = '%1.1f%%',
        shadow = True)
plt.title(name)
plt.legend()
plt.show()
```
